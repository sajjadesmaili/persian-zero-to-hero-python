```python
import statistics as ss

data2 = np.random.uniform(100,200,500)
means = ss.mean(data2)
print(means)
# 149.34396291183953

std = ss.stdev(data2)
print(std)
```

```python
data_new = (data2 - means )/ std
```

```python
print(min(data_new) , max(data_new) , ss.mean(data_new) , ss.stdev(data_new))
```
